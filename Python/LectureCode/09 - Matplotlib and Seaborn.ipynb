```python
# We already have a bit of experience with matplotlib and seaborn as Pandas uses them to do visualizations
# .plot() method calls matplotlib and you will need it installed

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# We can generate a series of numbers using np.linspace()
x = np.linspace(-10,10,100)

# We can create a sin wave using these points
y = np.sin(x)

# We can use matplotlib to generate any plot we want
plt.plot(x,y,marker="*",c="purple")
plt.show()
```

```python
df = pd.read_csv("../Data/california_housing_train.csv")

x = df["median_income"]
y = df["median_house_value"]

plt.scatter(x, y, marker="D", edgecolors="Black", c="Brown")
plt.show()
```

```python
# Let's take a look categorical data

education = ["High School", "Some College", "Bachelors", "Masters"]
values = [200, 130, 80, 30]
values2 = [30, 192, 38, 0]

# We can use subplots to create multiple visualizations
figs, axs = plt.subplots(3, 1)

axs[0].bar(education, values)
axs[1].scatter(education, values)
axs[2].plot(education, values2)

# Give the superplot a title
plt.suptitle("Education Data")
plt.show()
```

```python
# Seaborn

# Built on Matplotlib
# "Python" of visualizations

import seaborn as sns
```

```python
# Seaborn comes with built in themes and datasets

# Set the theme as default
sns.set_theme()

# load in a default dataset

tips = sns.load_dataset("tips")

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='smoker',
    style='sex',
    size='size'
    )
```
